```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as scc
import pandas as pd
import math
import scipy 
```

```python
fela, fexc, fion = [], [], []
with open("ncoll.dat") as f:
    line = f.readlines()
    for i, list in enumerate(line):
        item = list.split(" ")
        fela.append(int(item[0]))
        fexc.append(int(item[1]))
        fion.append(int(item[2]))
fela = np.array(fela)
fexc = np.array(fexc)
fion = np.array(fion)
nela, nexc, nion = fela/50000, fexc/50000, fion/50000
```

```python
freq = np.array([1.03380525e-01, 2.63223889e-04, 9.63823200e-05])

ind = np.arange(0,fela.shape[0])
ifela = (ind*0+1) * freq[0]
ifexc = (ind*0+1) * freq[1]
ifion = (ind*0+1) * freq[2]
plt.plot(ind, ifela, linewidth=2.5,color='r')
plt.plot(ind, (ind*0+1)*np.mean(nela), linewidth=2.5,color='b')
plt.scatter(ind, nela)
```

```python
plt.plot(ind, ifexc,linewidth=2.5,color='r')
plt.scatter(ind, nexc)
plt.plot(ind, (ind*0+1)*np.mean(nexc), linewidth=2.5,color='b')
```

```python
plt.plot(ind, ifion,linewidth=2.5,color='r')
plt.scatter(ind, nion)
plt.plot(ind, (ind*0+1)*np.mean(nion),linewidth=2.5,color='b')
```

```python
plt.scatter(ind,nela+nexc+nion)
plt.plot(ind, np.sum(freq)*(ind*0+1))
```
